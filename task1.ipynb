##### 6.clip和replace

### 7.apply函数  apply是一个自由度很高的函数  对于Series,他可以迭代每一列的值操作

```python
df['Math'].apply(lambda x:str(x)+'!').head()
```

对于DataFrame，他可以迭代每一个列操作：

```python
df.apply(lambda x:x.apply(lambda x:str(x)+'!')).head()
```
